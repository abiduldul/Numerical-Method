```python
import numpy as np

# ============================================================
# STEP 1: Input your 3 linear equations below
# Format: [a, b, c, d]  -->  ax + by + cz = d
# ============================================================

eq1 = [ 2,  1, -1,   8]   #  2x +  y -  z =  8
eq2 = [-3, -1,  2, -11]   # -3x -  y + 2z = -11
eq3 = [-2,  1,  2,  -3]   # -2x +  y + 2z = -3

# Build the augmented matrix [A | b]
matrix = np.array([eq1, eq2, eq3], dtype=float)

print('=' * 47)
print('         AUGMENTED MATRIX [A | b]')
print('=' * 47)
print(f"  {'':>4} {'x':>7} {'y':>7} {'z':>7}  {'| d':>7}")
print('-' * 47)
for i, row in enumerate(matrix):
    print(f"  eq{i+1}: {row[0]:>6.2f}  {row[1]:>6.2f}  {row[2]:>6.2f}  | {row[3]:>6.2f}")
print('=' * 47)

# ============================================================
# STEP 2: Forward Elimination (with Partial Pivoting)
# ============================================================

n = 3
aug = matrix.copy()

print('=' * 47)
print('           FORWARD ELIMINATION')
print('=' * 47)

for col in range(n):
    # Partial Pivoting: find row with max absolute value in current column
    max_row = col + np.argmax(np.abs(aug[col:, col]))
    if max_row != col:
        aug[[col, max_row]] = aug[[max_row, col]]
        print(f"  Swapped row {col+1} <-> row {max_row+1} (partial pivoting)")

    pivot = aug[col, col]
    if pivot == 0:
        raise ValueError(f"Zero pivot at column {col+1}. No unique solution.")

    print(f"\n  Pivot on column {col+1}:  pivot = {pivot:.4f}")

    for row in range(col + 1, n):
        factor = aug[row, col] / pivot
        aug[row] = aug[row] - factor * aug[col]
        print(f"    R{row+1} <- R{row+1} - ({factor:.4f}) x R{col+1}")

print('\n' + '=' * 47)
print('          UPPER TRIANGULAR MATRIX')
print('=' * 47)
for row in aug:
    print(f"  {row[0]:>8.4f}  {row[1]:>8.4f}  {row[2]:>8.4f}  | {row[3]:>8.4f}")
print('=' * 47)

# ============================================================
# STEP 3: Back Substitution
# ============================================================

x = np.zeros(n)

print('=' * 47)
print('             BACK SUBSTITUTION')
print('=' * 47)

for i in range(n - 1, -1, -1):
    x[i] = aug[i, n]
    for j in range(i + 1, n):
        x[i] -= aug[i, j] * x[j]
    x[i] /= aug[i, i]
    var = ['x', 'y', 'z'][i]
    print(f"  Solving for {var}: {var} = {x[i]:.6f}")

print('=' * 47)

# ============================================================
# STEP 4: Final Solution & Verification
# ============================================================

print('=' * 47)
print('               SOLUTION')
print('=' * 47)
print(f"  x = {x[0]:.6f}")
print(f"  y = {x[1]:.6f}")
print(f"  z = {x[2]:.6f}")
print('=' * 47)
```
